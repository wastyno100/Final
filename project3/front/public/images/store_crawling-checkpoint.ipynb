```python
#드라이버 지정=크롬
driver=webdriver.Chrome()

#자동실행

#목표 홈페이지 열기
driver.get("https://tpirates.com/") #홈페이지 열기
time.sleep(5) #로딩시간 5초(에러발생시 시간 늘려서 재실행해보기)

#시세 클릭
mall_click = driver.find_elements("css selector","#wrap > div.sc-eBMEME.cXOJzI > div.search-form > div.sc-dhKdcB.sc-cWSHoV.fsycso.dqpzTc >div:nth-child(3) > a")[0]
mall_click.click() #클릭
 #로딩시간 3초(에러발생시 시간 늘려서 재실행해보기)
```

```python
from time import localtime
from time import strftime
from time import time
data_list = []

for i in range(1, 200):
   
    # 10의 배수인 경우 스크롤 내리고 10초 대기
    if i % 10 == 1 and i > 1:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")  
        driver.implicitly_wait(30)
        #print('inner if')
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
    
    try:
        num=i
        
        # 상품명
        name = soup.select(f"ul.sc-xwuxA.isTdUO > li:nth-child({i}) > div > div >div.sc-eeDRCY.kPzmFC")[0].text

        # 가격
        extracted_number = soup.select(f"ul.sc-xwuxA.isTdUO > li:nth-child({i}) > div > div >div.sc-eeDRCY.cMAA-Df")[0].text

        # 숫자변환 정규표현식
        formatted_number = re.sub(r'[^\d]', '', extracted_number)
        price = int(formatted_number)

        tm = time()
        today = localtime(tm)
        day = strftime('%Y-%m-%d', today)
        data_list.append([num,name,price,day])        
        #print("inner try")
        
    except Exception as e:
        pass
```
